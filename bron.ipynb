```python
namespace_name = "ehns-clickstream-vastav"
eventhub_name = "clickstream-events"
connection_string = "<YOUR_EVENT_HUB_CONNECTION_STRING>"  # store securely, e.g. in Databricks secrets or env var"

bootstrap_servers = f"{namespace_name}.servicebus.windows.net:9093"

kafka_options = {
    "kafka.bootstrap.servers": bootstrap_servers,
    "subscribe": eventhub_name,
    "kafka.sasl.mechanism": "PLAIN",
    "kafka.security.protocol": "SASL_SSL",
    "kafka.sasl.jaas.config": f'kafkashaded.org.apache.kafka.common.security.plain.PlainLoginModule required username="$ConnectionString" password="{connection_string}";',
    "startingOffsets": "earliest"
}
```

```python
df = spark.readStream \
    .format("kafka") \
    .options(**kafka_options) \
    .load()

display(df)
```

```python
checkpoint_path = "abfss://bronze@itsmystorage.dfs.core.windows.net/clickstream_catalog/_checkpoints/bronze_raw_display/"

display(df, checkpointLocation=checkpoint_path)
```

```python
from pyspark.sql.functions import col

# Cast the Kafka value (binary) to string — this is the raw JSON event
bronze_df = df.select(
    col("value").cast("string").alias("raw_event_json"),
    col("topic"),
    col("partition"),
    col("offset"),
    col("timestamp").alias("kafka_timestamp")
)
```

```python
bronze_checkpoint = "abfss://bronze@itsmystorage.dfs.core.windows.net/clickstream_catalog/_checkpoints/bronze_write/"

bronze_query = bronze_df.writeStream \
    .format("delta") \
    .option("checkpointLocation", bronze_checkpoint) \
    .outputMode("append") \
    .toTable("clickstream_catalog.bronze.raw_events")
```
